```python

# 練習
chinese = 78
english = 93
math = 73

total = chinese + english + math
print("total=", total)
```

```python
# 字串和變數的整合
apple = 5
orange = 9

print("蘋果=" + str(apple) + "個,橘子=" + str(orange) + "個")
```

```python
# str template
# 字串插補變數 f
# f的功能是將字串和變數整合,傳出1個全新的字串
print(f"蘋果={apple}個,橘子={orange}個")
```

```python
# 練習 指定變數，將字串和變數放在一起傳出新的字串
value1 = 3
value2 = 18

print(f"生日={value1}月{value2}日" )
```

```python
# input()
# input的傳出值是字串
# 使用int將input的傳出值變為整數
apple = int(input("請輸入蘋果的數量:"))
type(apple)
```

```python
# 練習使用input
value3 = input("請輸入轉帳金額")
print(int(value3))
```
