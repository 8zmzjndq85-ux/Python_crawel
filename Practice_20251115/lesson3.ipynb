```python
# boolean 布林值：只有true和false
# if 判斷，傳出布林值
# if + 運算式
# if + 比較運算式 → 單項，傳出值為True才執行
# if + 比較運算式 + else → 雙項，傳出值若為true, 執行if 部分；傳出值若為false，執行else部分
```

```python
value4= int(input("請輸入年齡"))

if value4 >= 18:
    print("交女朋友要看身分證")

else:
    print("不要交女朋友")
```

```python
#巢狀判斷
value4= int(input("請輸入年齡"))

if value4 >= 20:
    print("可以投票")

else:
    if value4 >=18:
        print("民法未成年")
    else:
        print("不要交女朋友")
    
```

```python
#多項選擇elif: 將else與if合併
value4= int(input("請輸入年齡"))

if value4 >= 20:
    print("可以投票")

elif value4 >=18:
    print("民法未成年")
else:
    print("不要交女朋友")
```

```python
score = 59
if score >= 90:
    print("優")
elif score >= 80:
    print("甲")
elif score >= 70:
    print("乙")
elif score >= 60:
    print("丙")
else:
    print("丁")
```

```python
# and: 嚴格的得到True, and兩側的運算元都傳出True，該行才會傳出True;
# or: 容易的得到True, and兩側的運算元有一個傳出True，該行就會得到True
score = 95
if score >= 90:
    print("優")
elif score< 90 and score >= 80:
    print("甲")
elif score < 80 and score >= 70:
    print("乙")
elif score < 70 and score >= 60:
    print("丙")
else:
    print("丁")
```

```python
# 工具箱random
# randint(a,b)回傳隨機值介於a與b之間 a<=隨機值<=b
# while + 布林值運算式 迴圈：若傳出值為真，執行指令程式碼，並再次執行布林值運算式；直到傳出值為False，跳出迴圈。  在不確定運算次數使用

import random


while random.randint(1,10) != 5:
    print("這不是5")

print("這是5")
```

```python
# while 後可以指定布林值(bool) ex. while True
# 在while的程式區可以使用break → 跳出迴圈；使用continue → 後面的程式區省略，直接進入下一次迴圈
import random


while True:
    rand_value = random.randint(1,10)
    if rand_value == 5:
        print("這是5")
        break
    print("這不是5")
```
