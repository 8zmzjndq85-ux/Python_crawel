```python
import time 

def download_file(file_name): # file_name 是參數
    print(f"開始下載 {file_name}...")
    time.sleep(2)  # 模擬下載文件所需的時間
    print(f"{file_name} 下載完成！")
    

def main_sync():
    start_time = time.time()
    download_file("檔案A.pdf") # "檔案A.pdf" 是引數
    download_file("檔案B.pdf")
    download_file("檔案C.pdf")
    end_time = time.time()
    print(f"花費的時間: {(end_time - start_time):.2f} 秒")

if __name__ == "__main__":
    main_sync()
```

```python
#範例：在if __name__ == "__main__":使用asyncio建立事件迴圈
#Function要加上async才能在事件迴圈執行；async def` 定義的函式才能在事件迴圈中執行，且才能用 await。
#await是將任務丟進事件迴圈執行，但是一次只能丟一個，await只能放在async內；而且await只能執行async function，因此download_file這個function前面要加上async
#`await` 只能在 `async` 函式內使用，且只能等待一個任務。
#使用asyncio的函式gather將三個任務同時丟進事件迴圈執行；若要同時執行多個任務，需用 `await asyncio.gather(...)`，這樣三個任務會同時進行。
#`asyncio.run(main_async())` 會在 `if __name__ == "__main__":` 啟動事件迴圈。
import asyncio
import nest_asyncio

nest_asyncio.apply()

async def download_file(file_name):
    print(f"開始下載 {file_name}...")
    await asyncio.sleep(2)  # 模擬下載文件所需的時間
    print(f"{file_name} 下載完成！")

async def main_async():
    start_time = time.time()
    await asyncio.gather(
        download_file("檔案A.pdf"),
        download_file("檔案B.pdf"),
        download_file("檔案C.pdf")
    )
    end_time = time.time()
    print(f"花費的時間: {(end_time - start_time):.2f} 秒")

if __name__ == "__main__":
    asyncio.run(main_async())
```
